#Movie Recommendation Using Machine Learning


importing the dependencies


```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

data collection and pre-processing


```python
#loading the data from csv file to panda datafile
movies=pd.read_csv('/content/movies.csv')
```

```python
#print first 8 rows
movies.head(8)
```

```python
# finding number of rows and columns
movies.shape
```

```python
#feature selction
selected_feature=['genres', 'keywords', 'tagline', 'cast', 'director'	]
print(selected_feature)
```

```python
#replacing the null value with null string
for feature in selected_feature:
  movies[feature]=movies[feature].fillna('')
```

```python
#combining all the selected features
combined_features=movies['genres']+''+movies['keywords']+''+movies['tagline']+''+movies['cast']+''+movies['director']
```

```python
print(combined_features)
```

```python
#converting the text intofeature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

```python
# similarity score using cosine similarity function
similarity=cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the movie name from user
movie_name=input("enter your favorite movie:")
```

```python
#creating a list of all name of movie in dataset
list_of_movie=movies['title'].tolist()
print(list_of_movie)
```

```python
find_movie_name=difflib.get_close_matches(movie_name,list_of_movie)
print(find_movie_name)
```

```python
close_match=find_movie_name[0]
print(close_match)
```

```python
#finding the index of the movie with title
index_of_movie=movies[movies.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
#getting list of similar movies
similarity_score=list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on thier score
sorting_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorting_similar_movies)
```

```python
#print the name of similar movies based on index
print('Movies recomended for you: \n')
i=1
for movie in sorting_similar_movies:
  index=movie[0]
  title_from_index=movies[movies.index==index]['title'].values[0]
  if(i<10):
    print(i,'.',title_from_index)
    i+=1
```

```python
#Checking the whole in one cell
movie_name=input("enter your favorite movie:")
list_of_movie=movies['title'].tolist()
find_movie_name=difflib.get_close_matches(movie_name,list_of_movie)
close_match=find_movie_name[0]
index_of_movie=movies[movies.title == close_match]['index'].values[0]
similarity_score=list(enumerate(similarity[index_of_movie]))
sorting_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print('Movies recomended for you: \n')
i=1
for movie in sorting_similar_movies:
  index=movie[0]
  title_from_index=movies[movies.index==index]['title'].values[0]
  if(i<10):
    print(i,'.',title_from_index)
    i+=1
```
